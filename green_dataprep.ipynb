```python
# Set timeframe to "1Day"
timeframe = "1Day"

# Set start and end datetimes of 1 year, between now and 365 days ago.
start_date = pd.Timestamp("2019-10-03", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2023-10-06", tz="America/New_York").isoformat()

# Set the stock tickers
tickers = ["VOO", "QCLN"]

# Get 3 year's worth of historical data for all stocks
index_df = api.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date, limit = 10000
).df

# Display sample data
index_df.head(50)
```

```python
# Choose relevant columns
index_df = index_df [
    ["close", "symbol"]
]
index_df
```

```python
# Drop time from timestamp column
# Define a mapping of old column names to new column names
column_mapping = {
    'timestamp' : 'date',
}

# Rename the columns in the DataFrame
index_df.rename(columns=column_mapping, inplace=True)

# Your DataFrame now has the desired column names
print(index_df)
```
